**LOADING DATA IN PANDA**

```python
##library 
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
import time
from bs4 import BeautifulSoup
import requests
%matplotlib inline
##loading the data

df=pd.read_csv('Netflix Shows.csv', encoding='latin-1')
df
```

**READ THE DATA**

```python
df.head(100)
```

**Drop duplicate rows**


```python
nf=df.drop_duplicates()
nf
```

**DESCRIBING THE DATA**

```python
nf.describe()
```

**SORTING**

```python
nf.sort_values('release year',ascending=False)
```

**AGGREGATE STATISTICS (GROUPBY)**

```python
df.groupby(['title']).mean().sort_values('user rating size',ascending=False)
```

**FILTERING THE DATA**

```python
new_nf=nf.loc[df['user rating size']>80]
new_nf
```

**BAR GRAPH**

```python
score_freq = nf['user rating size'].value_counts()

plt.figure(figsize=(15,10))
score_freq.plot.bar(color='red', alpha=0.5)
plt.xticks(rotation=50)
plt.xlabel("User Rating size")
plt.ylabel("Frequency of user rating size")
plt.title("User Rating Size Frequency")
plt.show()
```
